LOADING REQUIRED LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
df = pd.read_csv(r'E:\Data Cleaning\AB_NYC_2019.csv')
print(df.head())
```

MISSING DATA HANDLING

```python
df_cleaned = df.dropna()
```

DUPLICATE REMOVAL

```python
df_cleaned = df_cleaned.drop_duplicates()
```

STANDARDIZATION

```python
df_cleaned['name'] = df_cleaned['name'].str.lower()
```

```python
Q1 = df_cleaned['host_id'].quantile(0.25)
Q3 = df_cleaned['host_id'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df_cleaned[(df_cleaned['host_id'] >= lower_bound) & (df_cleaned['host_id'] <= upper_bound)]
```

UPLOADING IMAGE

```python
image = Image.open(r'E:\Data Cleaning\New_York_City_.png')
```

VISUALIZE THE IMAGE

```python
plt.imshow(image)
plt.axis('off')
plt.show()
```
